## Второй коллоквиум

Второй коллоквиум для экономии времени пройдет в письменном формате. Вам выдается листок с двумя случайным образом выбранными вопросами из тех, которые ниже

Вы должны написать связный подробный теоретический ответ на вопрос

### Примеры вопросов по теме **"Деревья решений и ансамбли"**  

- Опишите алгоритм построения дерева решений. Какие основные шаги он включает?
- Что такое критерий информативности? Как он работает при простроении деревьев?
- Какие есть критерии останова у построения дерева? Почему именно такие?
- В чем разница между беггингом и бустингом?
- Что такое стекинг?
